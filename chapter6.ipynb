### Translation

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png').convert('L')
img1 = scipy.misc.fromimage(img)

plt.imshow(img)
plt.show()

# translate by 10 pixels in x and 4 pixels in y

transformation = AffineTransform(translation=(10, 4))
# nearest neighbor order = 0
img2 = warp(img1, transformation)
im4 = scipy.misc.toimage(img2).convert('LA')
# img = 
plt.imshow(im4)
plt.show()

transformation = AffineTransform(translation=(4, 10))
# nearest neighbor order = 0
img21 = warp(img1, transformation)
im41 = scipy.misc.toimage(img21).convert('LA')
# img = 
plt.imshow(im41)
plt.show()
```

#### 6.2.2 Rotation

```python
import numpy as np
import scipy.misc,math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png')
img1 = scipy.misc.fromimage(img)
plt.imshow(img)
plt.show()

# rotation angle in radians
transformation = AffineTransform(rotation=-0.5)
img2 = warp(img1, transformation)
# img4 = scipy.misc.toimage(img2).convert('LA')
plt.imshow(img2)
plt.show()

transformation = AffineTransform(rotation=1)
img21 = warp(img1, transformation)
# img4 = scipy.misc.toimage(img2).convert('LA')
plt.imshow(img21)
plt.show()
```

#### 6.2.3 Scaling

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png').convert('L')
img1 = scipy.misc.fromimage(img)
plt.imshow(img)
plt.show()
#scale by1/2 on both x and y.
transformation = AffineTransform(scale= (0.5, 0.5))
img2 = warp(img1, transformation)
im4 = scipy.misc.toimage(img2).convert('LA')
plt.imshow(im4)
plt.show()

transformation = AffineTransform(scale= (2, 2))
img21 = warp(img1, transformation)
im41 = scipy.misc.toimage(img21).convert('LA')
plt.imshow(im41)
plt.show()
```

#### 6.2.4 Interpolation

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png')
img1 = scipy.misc.fromimage(img)
plt.imshow(img)
plt.show()
transformation = AffineTransform(scale=(0.1, 0.1))

# nearest neighbor order = 0
img2 = warp(img1, transformation, order=0)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

# bi-linear order = 1
img2 = warp(img1, transformation, order=1) # default
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#bi-quadratic order = 2
img2 = warp(img1, transformation, order=2)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#bi-cubic order = 3
img2 = warp(img1, transformation, order=3)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()
```
