```python
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image using cv2.
a = cv2.imread('../Figures/ultrasound_muscle.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)

# Initializing the filter of size 10 by 10.
# The filter is divided by 10000 for normalization.
m = np.ones((10,10))/100
# Initializing the filter of size 100 by 100.
# The filter is divided by 10000 for normalization.
k = np.ones((100,100))/10000
# performing convolution
b = scipy.ndimage.filters.convolve(a, k)
c = scipy.ndimage.filters.convolve(a, m)
# Writing a,b and c to a file.
plt.imshow(a)
plt.show()
plt.imshow(c)
plt.show()
plt.imshow(b)
plt.show()
```

#### Median Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('../Figures/ct_saltandpepper.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing the median filter.
b = scipy.ndimage.filters.median_filter(a, size=2)
c = scipy.ndimage.filters.median_filter(a, size=10)
d = scipy.ndimage.filters.median_filter(a, size=100)
# Saving b as median_output.png in Figures folder
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
plt.imshow(d)
plt.show()
```

#### Max Filter 

```python
import scipy.misc
import scipy.ndimage
from scipy.misc.pilutil import Image
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
a = Image.open('../Figures/wave.png').convert('L')
# performing maximum filter
b = scipy.ndimage.filters.maximum_filter(a, size=2)
c = scipy.ndimage.filters.maximum_filter(a, size=20)
d = scipy.ndimage.filters.maximum_filter(a, size=100)
# b is converted from an ndarray to an image
b = scipy.misc.toimage(b)
c = scipy.misc.toimage(c)
d = scipy.misc.toimage(d)
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
```

#### Min Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
a = cv2.imread('../Figures/wave.png')

# performing minimum filter
b = scipy.ndimage.filters.minimum_filter(a, size=5)
c = scipy.ndimage.filters.minimum_filter(a, size=100)
d = scipy.ndimage.filters.minimum_filter(a, size=1000)
# saving b as mino.png
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
plt.imshow(d)
plt.show()
```

#### Sobel Filter

```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/cir.png')
# Converting a to grayscale .
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing Sobel filter.
b = ndimage.sobel(a)
# Saving b.
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
```

#### Canny Filter

```python
import cv2
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/maps1.png')
# Performing Canny edge filter.
b = cv2.Canny(a, 20, 30)
c = cv2.Canny(a, 2, 3)
d = cv2.Canny(a, 40, 60)
# Saving b,c and d.
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
plt.imshow(d)
plt.show()
```

#### Laplacian Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/imagefor_laplacian.png')
# Performing Laplacian filter.
b = scipy.ndimage.filters.laplace(a,mode='reflect')
c = scipy.ndimage.filters.laplace(a,mode='mirror')
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
```

# Laplacian of Gaussian Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/vhuman_t1.png')
# Performing Laplacian of Gaussian.
b = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='reflect')
c = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='constant')
plt.imshow(a) 
plt.show()
plt.imshow(b) 
plt.show()
plt.imshow(c) 
plt.show()
```

# Frangi Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
from PIL import Image
from skimage.filters import frangi

img = cv2.imread('../Figures/angiogram1.png')
img1 = np.asarray(img)
img2 = frangi(img1, black_ridges=True)
img3 = 255*(img2-np.min(img2))/(np.max(img2)-np.min(img2))

plt.imshow(img)
plt.show()
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
plt.imshow(img3)
plt.show()
```
